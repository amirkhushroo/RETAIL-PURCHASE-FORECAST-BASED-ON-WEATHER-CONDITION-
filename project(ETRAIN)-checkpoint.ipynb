<h2 style="font-family:Times New Roman;color:red;">LINE PLOTS</h2>

```python

plt.figure(figsize=(10, 5))
sns.lineplot(data=rsidf,    x='Weather Condition',    y='Units Sold',     marker='o',   markersize=8, linewidth=2, color='royalblue')
plt.title('Impact of Weather Conditions on Average Sales', pad=20, fontsize=14)
plt.xlabel('Weather Condition', fontsize=12)
plt.ylabel('Average Units Sold', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

<h3>Key Insights from This Plot:</h3>
<ul>
  <li><strong>Sunny days</strong> show the highest average sales.</li>
  <li><strong>Rainy weather</strong> has the lowest sales performance.</li>
  <li><strong>Cloudy</strong> and <strong>Snowy</strong> conditions show moderate sales impact.</li>
</ul>


```python
plt.figure(figsize=(14, 7))
sns.lineplot(data=rsidf, x='Price_Bin', y='Competitor Pricing', hue='Category',style='Category', markers=True,dashes=False, markersize=8, linewidth=2)
plt.title('Competitor Pricing Analysis by Category', fontsize=16)
plt.xlabel('Our Price Range ($)')
plt.ylabel('Competitor Price ($)')
plt.legend(bbox_to_anchor=(1.05, 1))
plt.grid(axis='y', alpha=0.3)
plt.show()
```

```python

# Calculate average units sold by season
seasonal_sales = rsidf.groupby('Seasonality')['Units Sold'].mean().reset_index()
# Create plot
plt.figure(figsize=(12, 6))
ax = sns.lineplot(data=seasonal_sales, x='Seasonality',y='Units Sold',marker='o', markersize=10,linewidth=2.5,
                 color='#3498db')

# Customize plot
plt.title('Seasonal Sales Performance', fontsize=16, pad=20)
plt.xlabel('Season', fontsize=12)
plt.ylabel('Average Units Sold', fontsize=12)
plt.ylim(0, seasonal_sales['Units Sold'].max() * 1.1)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

<h1 style="color:red;">BAR PLOTs</h1>


```python
# Get unique store IDs
store_ids = rsidf['Store ID'].unique()

# Loop through each store and plot
for store in store_ids:
    plt.figure(figsize=(10, 5))
    store_df = rsidf[rsidf['Store ID'] == store]
    
    sns.barplot(x='Weather Condition', y='Units Sold', hue='Category', data=store_df, estimator='mean')
    plt.yscale('log')
    plt.title(f"Store ID: {store} - Category vs Weather")
    plt.xticks(rotation=45)
    plt.tight_layout()
    
    plt.show()
```

### key insights
Sunny weather generally drives higher sales across multiple categories.

Rainy and Cloudy days may see lower sales, especially in less essential categories.

Certain categories like Groceries tend to maintain steadier sales across weather conditions, indicating essential demand.

```python
# Get unique regions
regions = rsidf['Region'].unique()

# Loop through each region and plot
for region in regions:
    plt.figure(figsize=(10, 5))
    region_df = rsidf[rsidf['Region'] == region]
    
    sns.barplot(x='Weather Condition', y='Units Sold', hue='Category', data=region_df, estimator='mean')
    plt.yscale('log')
    plt.title(f"Region: {region} - Category vs Weather")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
# Get unique product IDs
product_ids = rsidf['Product ID'].unique()

# Limit number of products to avoid too many plots (optional)
for product in product_ids[:5]:  # Show for first 5 products
    plt.figure(figsize=(10, 5))
    product_df = rsidf[rsidf['Product ID'] == product]
    
    sns.barplot(x='Weather Condition', y='Units Sold', data=product_df, estimator='mean')
    plt.yscale('log')
    plt.title(f"Product ID: {product} - Units Sold vs Weather")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
category_weather_df = rsidf.groupby(['Category', 'Weather Condition'])['Units Sold'].sum().reset_index()
plt.figure(figsize=(12, 6))
sns.barplot(x='Weather Condition', y='Units Sold', hue='Category', data=category_weather_df)
plt.yscale('log') 

plt.title('Units Sold by Category Across Weather Conditions')
plt.ylabel('Total Units Sold')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
product_season_df = rsidf.groupby(['Product ID', 'Seasonality'])['Units Sold'].sum().reset_index()
# Group the data by Product ID and Seasonality
product_season_df = rsidf.groupby(['Product ID', 'Seasonality'])['Units Sold'].sum().reset_index()

# Plot: grouped bar plot
plt.figure(figsize=(16, 7))
sns.barplot(x='Product ID', y='Units Sold', hue='Seasonality', data=product_season_df, palette='Set2')
plt.yscale('log')
plt.title('Units Sold per Product Across Seasons')
plt.xlabel('Product ID')
plt.ylabel('Total Units Sold')
plt.xticks(rotation=45)
plt.legend(title='Season')
plt.tight_layout()
plt.show()
```

<h2 style="font-family:Times New Roman;color:red;">HEAT MAP</h2>

```python
# Price Sensitivity by Category & Region
plt.figure(figsize=(14, 8))
price_bins = pd.qcut(rsidf['Price'], q=5)  # Create 5 price brackets
pivot = rsidf.pivot_table(values='Units Sold',
                         index='Category',
                         columns=price_bins,
                         aggfunc='sum',
                         fill_value=0)
sns.heatmap(pivot, cmap='coolwarm', annot=True, fmt=',.0f', 
           linewidths=0.5, linecolor='gray')
plt.title('Price Sensitivity Analysis\n(Units Sold by Price Range and Category)', 
         pad=20, fontsize=14)
plt.xlabel('Price Ranges', fontsize=12)
plt.ylabel('Product Category', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 7))
discount_bins = [0, 5, 10, 15, 20, 25]  # Custom discount brackets
pivot = rsidf.pivot_table(values='Units Sold', index='Category', columns=pd.cut(rsidf['Discount'], bins=discount_bins), aggfunc='mean', fill_value=0)                       
sns.heatmap(pivot, cmap='YlOrRd', annot=True, fmt='.0f',cbar_kws={'label': 'Average Units Sold'})
plt.title('Discount Effectiveness by Product Category', pad=20, fontsize=14) 
plt.xlabel('Discount Range (%)', fontsize=12)
plt.ylabel('')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
pivot = rsidf.pivot_table(values='Demand Forecast', index='Weather Condition',columns='Category',aggfunc='median',  # Using median to reduce outlier impact
fill_value=0)
sns.heatmap(pivot, cmap='viridis', annot=True, fmt='.0f', linewidths=0.3, annot_kws={'size':9})       
plt.title('Median Demand Forecast by Weather Condition', pad=15, fontsize=14) 
plt.xlabel('Product Category', fontsize=12)
plt.ylabel('')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
